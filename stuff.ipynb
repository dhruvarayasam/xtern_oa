```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
scaler = StandardScaler()
```

```python
v_data = pd.read_csv('data.csv')
v_data.drop_duplicates(inplace=True)
```

```python
# which users are most likely to purchase virtual merch?
# how do user activities during race events correlate w/ merch purchases + sponshorship interaction?
# a third research question

# which users are most likely to purchase virtual merch?
# test for users that interact with velocityx through Live360 more vs less
# test for users that complete challenges

X = v_data[['Sponsorship Interactions (Ad Clicks)', 'Virtual Merchandise Purchases']]
kmeans = KMeans(n_clusters=3, random_state=42)

kmeans.fit(X)

v_data['cluster'] = kmeans.predict(X)
plt.scatter(X['Sponsorship Interactions (Ad Clicks)'], X['Virtual Merchandise Purchases'], c=v_data['cluster'], cmap='viridis', marker='o')
plt.title('Clusters of Data')
plt.xlabel('Sponshorship Interactions (Ad Clicks)')
plt.ylabel('Merchandise Purchases')
plt.show()

score = silhouette_score(X, v_data['cluster'])
print(f'Silhouette Score: {score}')

```

```python
feature1 = 'Time on Live 360 (mins)'
feature2 = 'Virtual Merchandise Purchases'

X = v_data[[feature1, feature2]]
kmeans = KMeans(n_clusters=3, random_state=42)

kmeans.fit(X)

v_data['cluster'] = kmeans.predict(X)
plt.scatter(X[feature1], X[feature2], c=v_data['cluster'], cmap='viridis', marker='o')
plt.title('Clusters of Data')
plt.xlabel(feature1)
plt.ylabel(feature2)
plt.show()

score = silhouette_score(X, v_data['cluster'])
print(f'Silhouette Score: {score}')
```
